```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/titanic.csv')
df.head()
```

```python
fr = df['Survived'].value_counts()
fr.index = ['Non-survivors', 'Survivors']
fr
```

```python
df.describe(include='all')
```

```python
df['Fare'].var()
```

```python
df.mode()['Age']
```

```python
from scipy.stats import skew
data = df['Age'].dropna()
coef_assimetria = skew(data)

plt.figure(figsize=[10,6])
plt.hist(data,  bins='auto')
plt.grid(True)
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
coef_assimetria
```

```python
fare = df['Fare'].dropna()
df['Pclass'].dropna()
g = df[['Pclass','Fare']].corr()

plt.figure(figsize=[8,6])
plt.hist(fare,  bins='auto', color = 'blue')
plt.grid(True)
plt.xlabel('Fare x Class')
plt.ylabel('Frequência')
plt.show()
g
```

```python
df.groupby(by='Survived')['Fare'].mean()
```

```python
h = df[['Age','Fare']].cov()
print(df[['Age','Fare']].corr())
print(h)
```
